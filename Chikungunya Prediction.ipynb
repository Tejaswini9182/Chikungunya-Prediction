```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
df = pd.read_csv('/content/chikungunya.csv')

# Drop unnecessary columns
df = df.drop(columns=[df.columns[16], df.columns[17]])

# Separate features (x) and target (y)
x = df.drop('Severe Chikungunya', axis=1)
y = df['Severe Chikungunya']

# Replace 'yes' and 'no' with 1 and 0 in both x and y
x = x.replace({'yes': 1, 'no': 0})
y = y.replace({'yes': 1, 'no': 0})

# Split the data into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)

# Ensure train_x and test_x are DataFrames with correct columns
train_x = pd.DataFrame(train_x, columns=x.columns)
test_x = pd.DataFrame(test_x, columns=x.columns)

# Apply Label Encoding before fitting the model
encoder = LabelEncoder()
categorical_cols = ['sex', 'joint pains', 'myalgia', 'headache', 'fatigue',
                   'vomitting', 'arthritis', 'Conjuctivitis', 'Nausea',
                   'Maculopapular rash', 'Eye Pain', 'Chills', 'Swelling']

for col in categorical_cols:
    # Fit the encoder on combined data and transform train_x and test_x
    encoder.fit(pd.concat([train_x[col], test_x[col]]).astype(str))
    train_x[col] = encoder.transform(train_x[col].astype(str))
    test_x[col] = encoder.transform(test_x[col].astype(str))

# Create and train the Linear Regression model
model_lr = LinearRegression()
model_lr.fit(train_x, train_y)

# Make predictions
y_pred_train_lr = model_lr.predict(train_x)
y_pred_test_lr = model_lr.predict(test_x)

# Evaluate the model
train_mse_lr = mean_squared_error(train_y, y_pred_train_lr)
test_mse_lr = mean_squared_error(test_y, y_pred_test_lr)
train_r2_lr = r2_score(train_y, y_pred_train_lr)
test_r2_lr = r2_score(test_y, y_pred_test_lr)

# Print the evaluation metrics
print("Train MSE:", train_mse_lr)
print("Test MSE:", test_mse_lr)
print("Train R-squared:", train_r2_lr)
print("Test R-squared:", test_r2_lr)
```
